```python
import os
import pandas as pd

folder_path = '../data/Stocks/'
all_columns = set()

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path, nrows=0)  # Read only headers
        all_columns.update(df.columns)

print("Unique column names found:", all_columns)   
```

```python
import os
import pandas as pd

folder_path = '../data/Stocks/'
cols_to_remove = ['Percent Change', '% Change']  # Add any variant

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Convert column names to string and check for matches
        cols_to_drop = [col for col in df.columns if col in cols_to_remove]
        
        if cols_to_drop:
            df = df.drop(columns=cols_to_drop)
        
        # Convert 'Date' to datetime and sort
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.sort_values('Date').reset_index(drop=True)
        
        # Save back
        df.to_csv(file_path, index=False)
        print(f"Processed: {filename}, removed columns: {cols_to_drop}")   
```

```python
import os
import pandas as pd

folder_path = '../data/Stocks/'
consistent_name = 'Turnover'
columns_to_rename = ['Turn Over']  # Variants to standardize

for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)
        df = pd.read_csv(file_path)
        
        # Rename any variant to the consistent name
        for col in columns_to_rename:
            if col in df.columns:
                df = df.rename(columns={col: consistent_name})
                break  # Avoid multiple renames
        
        # Convert 'Date' to datetime and sort
        df['Date'] = pd.to_datetime(df['Date'])
        df = df.sort_values('Date').reset_index(drop=True)
        
        # Save back
        df.to_csv(file_path, index=False)
        print(f"Standardized columns in: {filename}")   
```
